```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
df = pd.read_csv('Top Indian Places to Visit.csv')
df
```

```python
col=['Type','Google review rating','Number of google review in lakhs','Entrance Fee in INR']
df=df[col]
df
```

```python
print(df.isna().sum())
```

```python
df=df.loc[df['Entrance Fee in INR']!=0]
```

```python
print(df.to_string())
```

```python
df=pd.get_dummies(df,dtype=int)
df
```

```python
print((df['Entrance Fee in INR']!=0).sum())
```

```python
X=df.drop(columns=['Entrance Fee in INR'])
Y=df['Entrance Fee in INR']
X
Y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
model=LinearRegression()
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.12)
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model2=Lasso(alpha=25,max_iter=100,tol=0.1)
model2.fit(x_train,y_train)
```

```python
model2.score(x_train,y_train)
```

```python
model2.score(x_test,y_test)
```

```python
model3=Ridge(alpha=25,max_iter=100,tol=0.1)
model3.fit(x_train,y_train)
model3.score(x_train,y_train)
```

```python
model3.score(x_test,y_test)
```
